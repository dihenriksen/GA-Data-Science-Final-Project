```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
# json_data = pd.read_json("foliohd_data.json")
open_json_data = open("foliohd_data.json")
json_data = json.load(open_json_data)
open_json_data.close()

json_data.keys()

json_data["features_by_type"].keys()

json_data["feature_labels"]

# df = pd.DataFrame()
# json_data["feature_labels"]
# whooops. I guess I don't have the actual event names here

# Each row of the dataframe should be a person
# Each column is an event type

open_event_names = open("event_type_names.json")
event_names = json.load(open_event_names)
open_event_names.close()

person_list = []

json_data["features_by_type"]["upgraded"].values()

upgraded_df = pd.DataFrame(json_data["features_by_type"]["upgraded"].values())
neutral_df = pd.DataFrame(json_data["features_by_type"]["neutral"].values())
churned_df = pd.DataFrame(json_data["features_by_type"]["churned"].values())

# A key for the prefixes in the event types
prefixes = {
            "tbfe_n": "before_first_session_count",
            "oec_n": "overall_event_count",
            "fsec_n": "first_session_event_count",
            "ssec_n": "subsequent_session_event_count"
            }
# Add a column indicating the known class for that person
# 0 => churned
# 1 => upgraded
# 2 => neutral (undecided)

c = list(np.zeros(len(churned_df)))
u = list(np.ones(len(upgraded_df)))
n = list(2*np.ones(len(neutral_df)))

churned_df["Result"] = c
neutral_df["Result"] = n
upgraded_df["Result"] = u

data_df = upgraded_df.append([neutral_df, churned_df])

# loop through the current column names
# calculate desired name for each current name
# push desired name onto list of new column names
# assign new column names to dataframe

new_columns = []
for col in churned_df.columns:
    if ":" in col: # check for column names with ids, which will have a colon
        first, second = col.split(":")
        name = event_names[second]
        new_col = first + ":" + name
        new_columns.append(new_col)
    else:
        new_columns.append(col)
        
churned_df.columns = new_columns

new_columns = []
for col in neutral_df.columns:
    if ":" in col: # check for column names with ids, which will have a colon
        first, second = col.split(":")
        name = event_names[second] # get readable name for the id
        new_col = first + ":" + name # create new column name
        new_columns.append(new_col)
    else:
        new_columns.append(col)
        
neutral_df.columns = new_columns

new_columns = []
for col in upgraded_df.columns:
    if ":" in col: # check for column names with ids, which will have a colon
        first, second = col.split(":")
        name = event_names[second] # get readable name for the id
        new_col = first + ":" + name # create new column name
        new_columns.append(new_col)
    else:
        new_columns.append(col)
        
upgraded_df.columns = new_columns
```

```python
data_df = upgraded_df.append(churned_df)

# 1. Fill nan values with 0. If it is nan, then they did not trigger that event.
# 2. Select only avg events, and events prefixed by oec_n, to do a simple
# analysis on merely total event count

all_columns = list(data_df.columns)

print prefixes
```

```python
columns = []

for i in all_columns:
    if (i.find("oec_n") == -1 and i.find("tbfe_n") == -1):
        columns.append(i)
        
for i in columns:
    print i
```

```python
split_count_data = data_df[columns]
split_count_data = split_count_data.fillna(0)
# <<<<<<
split_count_data.head()
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

features = list(split_count_data.columns)
features.remove("Result")

np_split_count_data = split_count_data.get(features).values

results = split_count_data.Result.values

# Split the data
split_count_train, split_count_test, results_train, results_test = train_test_split(
    np_split_count_data, results, test_size=0.20, random_state=0)

lr = LogisticRegression(C=1)
lr.fit(split_count_train, results_train)

results_predicted = lr.predict(split_count_test)

print "Accuracy: ", accuracy_score(results_test, results_predicted)
# <<<<<<
```

```python
split_count_correlations = (zip(lr.coef_[0], features))
split_count_correlations.sort()

sorted_correlation_values = [item[0] for item in split_count_correlations]
sorted_correlation_labels = [str(item[1]) for item in split_count_correlations]

x = np.arange(len(sorted_correlation_labels))
              
plt.bar(x, sorted_correlation_values)
plt.xticks(x + 0.5, sorted_correlation_labels, rotation=90)
plt.show()
# <<<<<<
```

```python
results_predicted_probabilities = lr.predict_proba(split_count_test)
print results_predicted_probabilities

tmp = []
for i in results_predicted_probabilities:
    tmp.append(i[0])

tmp = np.array(tmp)
print tmp.mean() # average degree of confidence in predictions
# <<<<<<
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

def plot_roc_curve(results_test, results_predicted_probabilities):
    fpr, tpr, thresholds = roc_curve(results_test, results_predicted_probabilities[:, 1])
    
    roc_auc = auc(fpr, tpr)
    # Plot ROC curve
    plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate or (1 - Specifity)')
    plt.ylabel('True Positive Rate or (Sensitivity)')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    
plot_roc_curve(results_test, results_predicted_probabilities)
# <<<<<<
```
